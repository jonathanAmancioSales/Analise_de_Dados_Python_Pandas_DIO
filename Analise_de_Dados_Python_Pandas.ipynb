## Custos, Receitas, Lucros e Descontos

```python
df['Valor_Custo'] = df['Custo_Unitário'].mul(df['Quantidade'])

df['Valor_Lucro'] = df['Valor_Venda'] - df['Valor_Custo']
```

```python
pd.options.display.float_format = '{:9,.2f}'.format
```

```python
print(f"Custo    total: R$ {df['Valor_Custo'].sum():8,.2f}")
print(f"Receita  total: R$ {df['Valor_Venda'].sum():8,.2f}")
print(f"Lucro    total: R$ {df['Valor_Lucro'].sum():8,.2f}")
print(f"Desconto total: R$ {df['Valor_Desconto'].sum():8,.2f}\n\n")

print('Custos, Receitas, Lucros, Descontos e Quantidade por Marca:', end='')
#display( df.groupby('Marca')['Valor_Venda'].sum().to_frame() )
display( df.groupby('Marca').agg({'Valor_Custo':'sum',
                                  'Valor_Venda':'sum',
                                  'Valor_Lucro':'sum',
                                  'Valor_Desconto':'sum',
                                  'Quantidade':'sum'}).reset_index() )
```

```python
#Proporções entre marcas:
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
df.groupby('Marca')[['Quantidade',
                     'Valor_Custo',
                     'Valor_Lucro']].sum().plot.pie(subplots=True,
                                    layout=(1,3), figsize=(18,6),
                                    title='Proporções de Custo, Lucro e Quantidade de produtos vendidos por Marca',
                                    legend=None, labeldistance=None,
                                    colors=colors, startangle=0,
                                    autopct='%.1f%%')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
#sns.violinplot(data=df, x='Produto', y='Valor_Venda', hue='Marca')
#plt.xticks(rotation=90);

plt.figure(figsize=(12,5))
plt.title('Valor do custo agrupado por marca');
sns.violinplot(data=df, x='Marca', y='Valor_Custo', cut=0);

plt.figure(figsize=(12,5))
plt.title('Valor da venda agrupado por marca');
sns.violinplot(data=df, x='Marca', y='Valor_Venda', cut=0);

plt.figure(figsize=(12,5))
plt.title('Valor do desconto sobre a venda agrupado por marca');
sns.violinplot(data=df, x='Marca', y='Valor_Desconto', cut=0);
```

```python
print('Custos, Receitas, Lucros, Descontos e Quantidade por Produto, ordenado por Lucro:')
display( df.groupby(['Marca',
                     'Produto']).agg({'Valor_Custo':'sum',
                                      'Valor_Venda':'sum',
                                      'Valor_Lucro':'sum',
                                      'Valor_Desconto':'sum',
                                      'Quantidade':'sum'}).reset_index().sort_values(by=['Valor_Lucro'],
                                                                                         ignore_index=True,
                                                                                         ascending=False) )
```

Note que, apesar de que a marca 'Contoso' seja líder quanto à quantidade de produtos vendidos, ela é a marca que obtem menos lucro.

```python
print('Custos, Receitas, Lucros, Descontos e Quantidade (de produtos vendidos) agrupados por Marca e por Loja:')
display( df.groupby(['Marca',
                     'ID_Loja']).agg({'Valor_Custo':'sum',
                                      'Valor_Venda':'sum',
                                      'Valor_Lucro':'sum',
                                      'Valor_Desconto':'sum',
                                      'Quantidade':'sum'}).reset_index().sort_values(by=['Valor_Lucro'],
                                                                                         ignore_index=True,
                                                                                         ascending=False) )
```

```python
print('Custos, Receitas, Lucros, Descontos e Quantidade (de produtos vendidos) por Loja:')
display( df.groupby(['ID_Loja']).agg({'Valor_Custo':'sum',
                                      'Valor_Venda':'sum',
                                      'Valor_Lucro':'sum',
                                      'Valor_Desconto':'sum',
                                      'Quantidade':'sum'}).reset_index().sort_values(by=['Valor_Lucro'],
                                                                                         ignore_index=True,
                                                                                         ascending=False) )
```

```python
df['Data_Venda_Ano'] =df['Data_Venda'].dt.year
df['Data_Venda_Mes'] =df['Data_Venda'].dt.month
df['Data_Venda_Week']=df['Data_Venda'].dt.week
```

```python
plt.figure(figsize=(15,5.5))
Ano=2008
sns.lineplot(data=df[df['Data_Venda_Ano']==Ano], x='Data_Venda_Mes', y='Valor_Lucro', hue='Marca',
             linewidth=1.0, linestyle='-', marker='o',  markersize=12)
plt.legend(loc='upper right')
plt.xlabel('Mês')
plt.ylabel('Lucro')
plt.title(f'Lucro por Mês no Ano de {Ano}, agrupado por Marca')
plt.show()
```
